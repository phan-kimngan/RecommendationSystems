```
import numpy as np
from keras.preprocessing.text import Tokenizer
from keras.preprocessing import sequence
from pyvi import ViTokenizer
import string
from google.colab import drive
import re
import os
import keras
from collections import OrderedDict
from keras.models import Sequential
from keras.layers import Dense, Embedding, BatchNormalization, LSTM, Dropout
drive.mount('/content/gdrive')
path = "gdrive/My Drive/Assignment/train/Da/"
```

```
def clean_document(doc):
    doc = ViTokenizer.tokenize(doc) #Pyvi Vitokenizer library
    doc = doc.lower() #Lower
    tokens = doc.split() #Split in_to words
    table = str.maketrans('', '', string.punctuation.replace("_", "")) #Remove all punctuation
    tokens = [w.translate(table) for w in tokens]
    tokens = [word for word in tokens if word]
    return tokens
```

```
import os
INPUT_LENGTH = 50
sequences = []
for f in os.listdir(path): 
    f1 = open(path+f, encoding='utf-16')
    doc = f1.read()
    print(doc)
    tokens = clean_document(doc)
    for i in range(INPUT_LENGTH + 1, len(tokens)):
        seq = tokens[i-INPUT_LENGTH-1:i]
        line = ' '.join(seq)
        sequences.append(line)
sequences
```

```
tokenizer = keras.preprocessing.text.Tokenizer(filters='!"#$%&()*+,-./:;<=>?@[\]^`{|}~ ')
tokenizer.fit_on_texts(sequences)
```

```
tokenizer.word_index
```

```
sequences_digit = tokenizer.texts_to_sequences(sequences)
sequences_digit 
```

```
vocab_size = len(tokenizer.word_index) + 1
```

```
import array 
sequences_digit = np.array(sequences_digit)
X, y = sequences_digit[:,:-1], sequences_digit[:,-1]
print(y)
y = keras.utils.to_categorical(y, num_classes=vocab_size)
print(y)
seq_length = X.shape[1]
```

```
model = Sequential()
model.add(Embedding(vocab_size, 50, input_length=50))
model.add(BatchNormalization())
model.add(LSTM(512, return_sequences=True))
model.add(LSTM(512))
model.add(Dense(100, activation='relu'))
model.add(Dropout(0.2))
model.add(BatchNormalization())
model.add(Dense(vocab_size, activation='softmax'))
model.summary()

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
model.fit(X, y, batch_size=512, epochs=20)
```

```
def preprocess_input(doc):
    tokens = clean_document(doc)
    print('Tokens của câu:',tokens)
    tokens = tokenizer.texts_to_sequences(tokens)
    tokens=np.array(tokens).ravel()
    zeros=np.zeros(50-len(tokens))
    tokens=np.append(zeros,tokens)
    #tokens = keras.preprocessing.sequence.pad_sequences(tokens, maxlen=50, truncating='pre')
    return np.reshape(tokens, (1,50))
```

```
def generate_text(text_input, n_words):
    tokens = preprocess_input(text_input)
    for _ in range(n_words):
        next_digit = model.predict_classes(tokens, verbose=0)
        tokens = np.append(tokens, next_digit)
        tokens = np.delete(tokens, 0)
        tokens = np.reshape(tokens, (1, 50))
    
    # Mapping to text  
    tokens = np.reshape(tokens, (50))
    out_word = []
    for token in tokens:
        for word, index in tokenizer.word_index.items():
            if index == token:
                out_word.append(word)
                break
    for i in range(len(out_word)-1):
      if out_word[i]==out_word[i+1]:
        out_word[i]=''
    return ' '.join(out_word)
```

```
seq = 'Để giải độc cho da mặt'
print('Câu tìm kiếm:',seq)
print('Kết quả dự đoán:',generate_text(text_input=seq, n_words=18))
```
